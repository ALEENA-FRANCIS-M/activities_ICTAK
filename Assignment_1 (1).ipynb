# 5.Perform meaningful visualizations using the dataset. Bring atleast 3 visualizations.

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='SL',y='SW',data=data,hue='Classification')
plt.title('Relationship between SL and SW for different Iris classifications')
plt.xlabel('SL')
plt.ylabel('SW')
```

```python
sns.countplot(data['Classification'])
plt.title('Count plot of Classification')
plt.xlabel('Classification',fontsize=12)
plt.ylabel('Count',fontsize=12)
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(data.corr(),annot=True)
plt.title('Heatmap on Correlation of features in the data',fontsize=14)
```

```python
sns.pairplot(data)
```

```python
plt.figure(figsize=(6,10))
sns.boxplot(x='Classification',y='PL',data=data)
plt.title('Effect of PL on Classification')
plt.xlabel('Classification')
plt.ylabel('PL')
```
