# 6. Do min max scaling on the feature set (Take ‘Survived’ as target).

```python
data.dtypes
```

```python
# doing dummy coding on 'Sex' and 'Embarked' columns.
d1 = pd.get_dummies(data,columns=['Sex','Embarked'])
d1
```

```python
y=d1['Survived']
x=d1.drop(['Survived','Name','Ticket'],axis=1)
x.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=.25)

from sklearn import linear_model
lr = linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)

from sklearn.metrics import mean_squared_error
print('MSE is ',mean_squared_error(y_test,predictions))

from sklearn.metrics import r2_score
print('R-squared value is ',r2_score(y_test,predictions))
```

```python
# Min max scaling
from sklearn import preprocessing
min_max = preprocessing.MinMaxScaler(feature_range=(0,1))
x=min_max.fit_transform(x)
x=pd.DataFrame(x)
```

```python
x.describe()
```

```python
# Checking MSE and R-squared values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=.25)

from sklearn import linear_model
lr = linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)

from sklearn.metrics import mean_squared_error
print('MSE is ',mean_squared_error(y_test,predictions))

from sklearn.metrics import r2_score
print('R-squared value is ',r2_score(y_test,predictions))
```

There is no change in the values of MSE and R-squared after Min max scaling.
