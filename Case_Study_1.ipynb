```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# 1. Read the dataset to the Python environment

```python
data = pd.read_csv('cars_data.csv')
data
```

# 2. Check for null values present in the dataset

```python
data.isna().sum()
```

No null value is present in the dataset.

# 3.Plot a bar graph of male vs female buyers participated in the sales

```python
plt.figure(figsize=(5,6))
sns.countplot(data['Buyer Gender'])
plt.title('Male vs Female Buyers')
plt.ylabel('Count',fontsize=12)
plt.xlabel('Buyer Gender',fontsize=12)
```

# 4. Find the top five cars based on their sales prices

```python
top_cars = data.sort_values('Sale Price',ascending=False)
top_cars.head()
```

Volkswagen Jetta III, Audi A6, Lexus LX, Mazda MX-6 and Lexus ES are the top five cars based on their sales price.

# 5. Find the least five cars based on their resell price

```python
least_cars=data.sort_values('Resell Price',ascending=True)
least_cars.head()
```

Chevrolet 1500, GMC Savana 1500, Jaguar XF, Mitsubishi 3000GT and Infiniti G37 are the least five cars based on their resell price.
