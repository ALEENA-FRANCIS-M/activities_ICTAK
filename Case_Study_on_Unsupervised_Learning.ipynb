# 2. Try out different clustering models in the wine dataset.
# 3. Find the optimum number of clusters in each model and create the model          with the optimum number of clusters.

### 1. K-Means Clustering

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
```

#### Finding optimum number of clusters

```python
# Elbow method
plt.plot(range(1,11),wcss,'o-')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Within-cluster sum of squares')
```

From the figure it can be clearly seen that the optimum number of clusters is 3.

```python
# Taking 3 as the number of clusters
kmeans = KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(df)    
type(df)

# Converting y_kmeans into a data frame
y_kmeans=pd.DataFrame(y_kmeans)
y_kmeans = y_kmeans.rename(columns={0:'label'})
```

```python
# Concatenating the given dataset and y_kmeans
df1 = pd.concat([data,y_kmeans], axis=1)
df1.head()
```

#### Visualizations with respect to K-Means Clustering

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(18,8))
sns.scatterplot(data=df1, x='Total_Phenols', y='Color_Intensity', hue='label', ax=ax[0])
sns.scatterplot(data=df1, x='Color_Intensity', y='Hue', hue='label', ax=ax[1])
```

1. The color intensity normally increased with the amount of total phenols, but when compared to the other two clusters, the color intensity in Cluster 0 is much high even for low values of Total_Phenols. Also, the amount of total phenols doesn't showed much effect on the color intensty in case of Cluster 2.
2. The value of Hue decreased with increase in color intensity. The color intensity is much higher and hue is much lower for Cluster 0 when compared to the other two clusters.

```python
sns.scatterplot(x=df1.index, y=df1['Alcohol'],hue=df1['label'])
plt.title('Alcohol content in different Clusters')
```

Wines in Cluster 1 have the highest amount of alcohol content followed by Cluster 0. Wines in Cluster 2 have the least alcohol content.

```python
sns.scatterplot(x=df1.index, y=df1['Magnesium'],hue=df1['label'])
plt.title('Amount of Magnesium in different Clusters')
```

A huge difference in magnesium amount cannot be seen among the clusters, but comparitively Cluster 2 has lower amounts of magnesium.

```python
sns.scatterplot(x=df1.index, y=df1['Proline'],hue=df1['label'])
plt.title('Proline in different Clusters')
```

The amount of Proline in Cluster 2 and Cluster 0 are almost same but Proline amount in Cluster 1 is much high compared to the other two.

### 2. Agglomerative Hierarchical Clustering

```python
# Dendrogram
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(df,method='ward'))
plt.title('Dendrogram')
```

From the dendrogram it is clear that the optimum number of clusters is 3.

```python
# Agglomerative Clustering
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
y_hc = hc.fit_predict(df)

# Converting y_hc into a data frame
y_hc=pd.DataFrame(y_hc)
y_hc = y_hc.rename(columns={0:'label'})
```

```python
# Concatenating the given dataset and y_hc
df2 = pd.concat([data,y_hc], axis=1)
df2.head()
```

#### Visualizations with respect to Agglomerative Hierarchical Clustering

```python
sns.scatterplot(x=df2.index, y=df2['Total_Phenols'],hue=df2['label'])
plt.title('Total Phenol content in different Clusters')
```

Wines in Cluster 2 have the highest total phenol content whereas those in Cluster 1 have the least total phenol content.

```python
sns.scatterplot(x=df2.index, y=df2['Malic_Acid'],hue=df2['label'])
plt.title('Amount of Malic Acid in different Clusters')
```

The amount of Malic acid in Cluster 1 varies over a range whereas the amount is comparitively less in Cluster 2.

```python
sns.scatterplot(x=df2.index, y=df2['Color_Intensity'],hue=df2['label'])
plt.title('Color Intensity in different Clusters')
```

Color intensity is comparitively less in Cluster 0.
